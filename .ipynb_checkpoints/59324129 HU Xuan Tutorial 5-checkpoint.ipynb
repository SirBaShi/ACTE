# Tutorial 5: Rule-Based Chatbot

This notebook uses a simple **rule-based chatbot** as a starting point and focuses on **one improvement task**:

## Task — Improve the Chatbot's User Experience
You will:
1. Make the chatbot ask a follow-up question after answering:  
   **“Is there anything else I can help you with?”**
2. Add simple “memory” so the chatbot can **repeat the previous response** if the user asks a related follow-up.


## Part A — Original version (run to observe behavior)

Run the cell below and try:
- `hi`
- `price product`
- `order status`

Notice: the chatbot answers, but it doesn’t explicitly ask if you need more help, and it doesn’t remember the last answer.


```python
# business_chatbot.py

def chatbot_response(user_input):
    # Convert user input to lowercase to handle case-insensitivity
    user_input = user_input.lower()

    # Define rules
    if "price" in user_input and "product" in user_input:
        return "The price of the product is $50."
    elif "order" in user_input and "status" in user_input:
        return "Your order is currently being processed and will be shipped in 3-5 business days."
    elif "customer" in user_input and "support" in user_input:
        return "You can reach our customer support at support@ourcompany.com."
    elif "return" in user_input and "policy" in user_input:
        return "You can return products within 30 days of purchase."
    elif "hello" in user_input or "hi" in user_input:
        return "Hello! How can I assist you today?"
    else:
        return "I'm sorry, I didn't understand that. Could you please rephrase?"

def chat():
    print("Welcome to the Business Chatbot! Ask me anything.")
    print("Type 'exit' to end the chat.")

    while True:
        user_input = input("You: ")

        if user_input.lower() == "exit":
            print("Goodbye! Have a great day!")
            break

        response = chatbot_response(user_input)
        print("Bot:", response)

# Run the chatbot
chat()
```

## Part B — Task (To Do): Improve the Chatbot's User Experience

### To Do checklist
1. Make the chatbot ask a follow-up question after answering:  
   **“Is there anything else I can help you with?”**
2. Add simple “memory” so the chatbot can **repeat the previous response** if the user asks a related follow-up.

```python
def chatbot_response(user_input, previous_response=None):
    user_input = user_input.lower()
    response = ""
    # TODO 1: Add the follow-up question
    follow_up_question = "Is there anything else I can help you with?"

    # TODO 2: Repeat previous response if input is related
    if previous_response and any(keyword in user_input for keyword in previous_response.split()):
        response = previous_response
    elif "price" in user_input and "product" in user_input:
        response = "The price of the product is $50."
    elif "order" in user_input and "status" in user_input:
        response = "Your order is currently being processed and will be shipped in 3-5 business days."
    elif "customer" in user_input and "support" in user_input:
        response = "You can reach our customer support at support@ourcompany.com."
    elif "return" in user_input and "policy" in user_input:
        response = "You can return products within 30 days of purchase."
    elif "hello" in user_input or "hi" in user_input:
        response = "Hello! How can I assist you today?"
    else:
        response = "I'm sorry, I didn't understand that. Could you please rephrase?"
        follow_up_question = None  # Don't ask follow-up if error

    # TODO 3: Return BOTH values
    return response, follow_up_question

def chat():
    print("Welcome to the Business Chatbot! Ask me anything.")
    print("Type 'exit' to end the chat.")
    # TODO 4: Add memory variable、
    previous_response = None
    
    while True:
        user_input = input("You: ").strip()
        if user_input.lower() == "exit":
            print("Goodbye! Have a great day!")
            break
        
        # TODO 5: Unpack the two returned values
        response, follow_up = chatbot_response(user_input, previous_response)
        
        # TODO 6: Print both response and follow-up
        print("Bot:", response)
        if follow_up:
            print("Bot:", follow_up)
        
        # TODO 7: Update memory
        previous_response = response

# Run the chatbot
chat()
```
